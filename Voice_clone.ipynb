```python
# Cell 2: Upload Audio
from google.colab import files
import os

print("Please upload your voice sample file...")
uploaded = files.upload()

# Store the path for the next step
input_audio_path = os.path.join("/content/", next(iter(uploaded)))
print(f"\n✅ File uploaded successfully: {os.path.basename(input_audio_path)}")
print("Great! Now move on to Step 3.")
```

```python
# Cell 3: User Prompt
#@title Enter the text for your cloned voice to speak
#@markdown ---
text_to_speak = "  King Leontes of Sicilia accuses his wife, Queen Hermione, of infidelity with his childhood friend, King Polixenes of Bohemia. Consumed by irrational jealousy, Leontes becomes obsessed with proving Hermione’s supposed adultery and orders her imprisonment before putting her on trial, where she steadfastly maintains her innocence.  Leontes’ baseless accusations lead to the destruction of his family: Hermione dies of grief, her newborn daughter is abandoned in a distant land, and her son, Mamillius, dies due to the distress caused by his father’s madness. Leontes’ abandoned daughter, Perdita, grows up as a shepherdess, unaware of her royal heritage. Sixteen years later, she returns to Sicilia.  This Standard Ebooks edition is based on William George Clark and William Aldis Wright’s 1887 Victoria edition, which is taken from the Globe edition." #@param {type:"string"}

print(f"✅ Text saved! The model will now say: '{text_to_speak}'")
print("\nReady to go! Run the final cell below to create the magic.")
```

```python
# Cell 4: The Complete Code (Corrected)
import os
import torch
import torchaudio as ta
from chatterbox.tts import ChatterboxTTS
from IPython.display import Audio, display

def run_voice_cloning():
    """
    Main function to handle the entire voice cloning process.
    """
    try:
        # --- 1. Input Validation ---
        if 'input_audio_path' not in globals() or 'text_to_speak' not in globals():
            raise NameError("Oops! Please make sure you have run Step 2 (Upload) and Step 3 (Enter Text) before this step.")

        print("▶️ Starting the voice cloning process...")

        # --- 2. Audio Preparation ---
        print("   - Preparing audio sample...")
        temp_wav_path = "/content/prompt.wav"
        waveform, sample_rate = ta.load(input_audio_path)
        # Resample to 24kHz, which is the optimal sample rate for Chatterbox
        resampler = ta.transforms.Resample(orig_freq=sample_rate, new_freq=24000)
        ta.save(temp_wav_path, resampler(waveform), 24000)

        # --- 3. Model Loading & Optimization ---
        device = "cuda" if torch.cuda.is_available() else "cpu"
        print(f"   - Loading Chatterbox model onto the {device}...")

        # Set environment variable to help with CUDA memory fragmentation
        os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

        # Load the model directly. The model is designed to handle its internal components.
        # Use .cuda() to move the entire model to GPU after loading.
        model = ChatterboxTTS.from_pretrained(device=device)

        if device == "cuda":
            print("   - Applying GPU optimizations (bfloat16) to T3 module...")
            # This is the primary module that benefits from bfloat16 to save VRAM
            # The 'conds3' attribute might not be directly exposed or is handled internally.
            model.t3 = model.t3.to(dtype=torch.bfloat16)
            # The .to(dtype=...) for model.conds3 is removed as it caused an AttributeError.
            # Chatterbox is designed to manage its conditional components internally.


        # --- 4. Speech Generation ---
        print(f"   - Generating speech...")
        generated_wav = model.generate(text_to_speak, audio_prompt_path=temp_wav_path)

        # --- 5. Save and Display Output ---
        output_path = "/content/cloned_voice_output.wav"
        # Move generated_wav to CPU before saving to avoid potential CUDA issues with torchaudio.save
        ta.save(output_path, generated_wav.cpu(), model.sr)

        print("\n🎉 Success! Your cloned voice is ready.")
        print("Listen to the result below. You can also download the file from the 'Files' panel on the left.")

        display(Audio(output_path, autoplay=False))

    except NameError as ne:
        print(f"❌ Setup Error: {ne}. Please ensure all previous cells (Steps 1, 2, and 3) have been run successfully.")
    except Exception as e:
        print(f"❌ An error occurred during processing: {e}")
        print("If this is a CUDA out of memory error, consider reducing the length of your text or trying again with a smaller prompt.")


# Run the main function
run_voice_cloning()
```
